```python
import pandas as pd
from sqlalchemy import create_engine
```

```python

# Database connection URL (Update with your PostgreSQL credentials)
db_url = "postgresql://postgres:suraj2535@localhost:5432/demo_sales"
engine = create_engine(db_url)
```

```python
# Load CSV file into a DataFrame
df = pd.read_csv("sales_data_sample.csv", encoding="ISO-8859-1")  # Make sure this is the correct path
df.head()
```

```python
df.info()
```

```python
# Rename columns to lowercase
df.columns = df.columns.str.lower()
df.info()
```

```python
# Convert date column to proper format
df['orderdate'] = pd.to_datetime(df['orderdate'])

# Ensure amount is a float
df['priceeach'] = df['priceeach'].astype(float)

# Load data into PostgreSQL
df.to_sql("sales_data", engine, if_exists="append", index=False)

print("CSV data successfully loaded into PostgreSQL!")
```
